```python
import pandas as pd

transactions = pd.read_csv('transactions.csv')
users = pd.read_csv('users.csv')
webinar = pd.read_csv('webinar.csv')
```

Во вложенных файлах находятся 3 таблицы. Необходимо решить задачу:

В таблице webinar.csv находится список посетителей вебинара, который прошел 1 апреля 2016 г.  
Нужно найти тех посетителей, кто впервые зарегистрировался в еЛаме (users.csv) после вебинара, и для каждого из них посчитать сумму его пополнений в системе (transactions.csv).  
Под посетителем вебинара мы понимаем один email

There are 3 tables in the attached files. The problem must be solved:

The webinar.csv table contains a list of visitors to the webinar that took place on April 1, 2016.  
We need to find those visitors who first registered in eLama (users.csv) after the webinar, and for each of them to calculate the amount of their deposits in the system (transactions.csv).  
By webinar visitor we mean one email.

```python
web_users = users.merge(webinar)
web_users
```

```python
pay_users = web_users[web_users['date_registration'] >= '2016-04-01']
pay_users
```

```python
superpay_users = pay_users.merge(transactions, how='right')
superpay_users
```

```python
group_users = superpay_users.groupby('email')['price'].sum()
group_users
```
